```python
import keras
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # Corrected import
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, BatchNormalization
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from keras.callbacks import ReduceLROnPlateau
import cv2
import os
import numpy as np
import pandas as pd
import gradio as gr

```

```python
labels = ['NORMAL', 'PNEUMONIA']
img_size = 150

def get_data(data_dir, labels):  # Added labels as a parameter
    data = []
    for label in labels:
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                if img_arr is not None:  # Ensure the image was read correctly
                    resized_arr = cv2.resize(img_arr, (img_size, img_size))
                    data.append([resized_arr, class_num])
            except Exception as e:
                print(f"Error reading image {img}: {e}")

    # Convert the list to a NumPy array with consistent shapes
    images, labels = zip(*data)  # Separate images and labels
    return np.array(images), np.array(labels)



```

```python
labels = ['NORMAL', 'PNEUMONIA']
train_images, train_labels = get_data('/content/drive/MyDrive/unzipped_folder/chest_xray/train', labels)
test_images, test_labels = get_data('/content/drive/MyDrive/unzipped_folder/chest_xray/test', labels)
val_images, val_labels = get_data('/content/drive/MyDrive/unzipped_folder/chest_xray/test', labels)
```

```python
# Initialize lists
x_train, y_train = [], []
x_val, y_val = [], []
x_test, y_test = [], []

# Populate training data
for feature, label in zip(train_images, train_labels):
    x_train.append(feature)
    y_train.append(label)

# Populate test data
for feature, label in zip(test_images, test_labels):
    x_test.append(feature)
    y_test.append(label)

# Populate validation data
for feature, label in zip(val_images, val_labels):
    x_val.append(feature)
    y_val.append(label)

```

```python
positives = []
negatives = []
for i in range(len(y_train)):
  if y_train[i]:
    positives.append(x_train[i])
  else:
    negatives.append(x_train[i])
```

```python
# Ensure x_train, x_val, and x_test are NumPy arrays before reshaping
x_train = np.array(x_train)
y_train = np.array(y_train)

x_val = np.array(x_val)
y_val = np.array(y_val)

x_test = np.array(x_test)
y_test = np.array(y_test)

# Now resize data for deep learning
x_train = x_train.reshape(-1, img_size, img_size, 1)
x_val = x_val.reshape(-1, img_size, img_size, 1)
x_test = x_test.reshape(-1, img_size, img_size, 1)




```

```python
x_test[0].shape
```

```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
y_val = y_val.reshape(-1,1)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=30,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)
datagen.fit(x_train)
```

```python


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dropout, Dense, BatchNormalization

model = Sequential()
model.add(Conv2D(32, (3, 3), strides=1, padding='same', activation='relu', input_shape=(150, 150, 1)))
model.add(BatchNormalization())
model.add(MaxPool2D((2, 2), strides=2, padding='same'))

model.add(Conv2D(64, (3, 3), strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(MaxPool2D((2, 2), strides=2, padding='same'))

model.add(Conv2D(128, (3, 3), strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPool2D((2, 2), strides=2, padding='same'))

model.add(Conv2D(256, (3, 3), strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(MaxPool2D((2, 2), strides=2, padding='same'))

model.add(Flatten())  # Corrected here
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

# Display the model summary
model.summary()

```

```python
model.compile(optimizer = "rmsprop" ,
            loss = "binary_crossentropy" ,
            metrics = ['accuracy'])
```

```python
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy',
                                            patience = 2,
                                            verbose=1,
                                            factor=0.3,
                                            min_lr=0.000001)
```

```python
history = model.fit(datagen.flow(x_train,y_train, batch_size = 32) ,
                    epochs = 10 ,
                    validation_data = datagen.flow(x_val,y_val) ,
                    callbacks = learning_rate_reduction)
```

```python

model.save_weights('major_project_pneumonia_detection.weights.h5')
```

```python
print("loss of the model is ----" , model.evaluate(x_test,y_test)[0])
print("Accuracy of the model is ----" , model.evaluate(x_test,y_test)[1]*100 , "%")
```

```python
epochs = list(range(10))
fig, ax = plt.subplots(1, 2)
train_acc = history.history['accuracy']
train_loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']
fig.set_size_inches(20, 10)

ax[0].plot(epochs, train_acc, 'go-', label='Training Accuracy')
ax[0].plot(epochs, val_acc, 'ro-', label='Validation Accuracy')
ax[0].set_title('Training & Validation Accuracy')
ax[0].legend()
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax[1].plot(epochs, train_loss, 'g-o', label='Training Loss')
ax[1].plot(epochs, val_loss, 'r-o', label='Validation Loss')  # Corrected label here
ax[1].set_title('Training & Validation Loss')
ax[1].legend()
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")

plt.show()
```

```python
def pneumoniaprediction(img):
    # Resize and convert to grayscale if necessary
    if img.shape != (150, 150):
        img = cv2.resize(img, (150, 150))

    if img.ndim == 3 and img.shape[2] == 3:  # If it's a color image
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    img = img.reshape(1, 150, 150, 1)  # Reshape to add batch dimension for the model
    prediction = model.predict(img)[0][0]  # Extract probability score

    confidence = round(prediction * 100, 2) if prediction > 0.5 else round((1 - prediction) * 100, 2)
    label = "Pneumonia" if prediction > 0.5 else "Normal"

    return f"{label} ({confidence}% confidence)"
```

```python
import gradio as gr

interface = gr.Interface(
    fn=pneumoniaprediction,  # Updated function with confidence score
    title="Pneumonia Detection using Chest X-Ray",
    description="Upload a Chest X-Ray image to detect whether the patient has Pneumonia or not. "
                "The model will also provide a confidence score for the prediction.",
    inputs=gr.Image(type="numpy"),  # Accepts images in NumPy format (compatible with OpenCV)
    outputs="text"  # Displays the classification with confidence
)

interface.launch(debug=True, share=True)

```
